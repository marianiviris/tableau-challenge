```python
DataBike = pd.concat([CB201810, CB201811, CB201812, CB201901, CB201902, CB201903, CB201904, CB201905, CB201905,
          CB201906, CB201907, CB201908, CB201909], ignore_index=True)

DataBike
```

```python
DataBike["User's Birth Year"].unique()
```

```python
#Drop the tail end of User's Birth Years 2 standard deviations below the mean
DataBike["User's Birth Year"].mean()-(2*DataBike["User's Birth Year"].std())
```

```python
DataBike = DataBike.drop(DataBike.index[(DataBike["User's Birth Year"] < 1961)])
DataBike
```

```python
#Calculate Age based on Birth Year
DataBike['Age'] = 2019 - DataBike["User's Birth Year"]
DataBike['Age'] = DataBike['Age'].astype(int)
DataBike.head()
```

```python
#Dropping Unknown (Gender=0) Genders
DCB = DataBike.drop(DataBike.index[(DataBike["User's Gender"] == 0)])
DCB.head()
```

```python
#Changing gender values from int to strings
DCB["User's Gender"] = DCB["User's Gender"].replace({1:"Male", 2:"Female"})
DCB.head()
```

```python
DCB.to_csv('DataCityBike.csv', encoding='utf-8', index=False)
```
